# Step 1: Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Explanation: pandas: for data handling

matplotlib and seaborn: for visualization

%matplotlib inline: ensures plots show inside your notebook

#  Step 2: Load the Dataset

```python
df = pd.read_csv("titanic.csv")

# Display the first 5 rows of the dataset
df.head()
```

#  Step 3: Understand the Data

```python
df.info()
print("\nGender counts:\n", df['Sex'].value_counts())

# Check how many missing values are in 'Age'
print("\nMissing Age values:", df['Age'].isnull().sum())
```

Sex is categorical

Age is numeric

Some Age values are missing (need to be handled)

# Step 4: Clean the Data

```python
df_clean = df.dropna(subset=['Age'])

# Check again
df_clean.info()
```

We're focusing only on rows with valid Age values so our histogram is accurate.

#  Step 5: Bar Chart – Gender Distribution

```python
# Plot the distribution of gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='Sex', palette='Set2')

# Add labels and title
plt.title("Gender Distribution on Titanic", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

This bar chart shows how many males vs females were in the dataset.

# Step 6: Histogram – Age Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df_clean, x='Age', bins=15, kde=True, color='skyblue')

# Add labels and title
plt.title("Age Distribution of Passengers", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows how the ages are distributed — KDE gives a smoothed density curve.

# Step 7 : Save Plots

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='Sex', palette='Set2')
plt.title("Gender Distribution on Titanic")
plt.savefig("gender_distribution.png")

plt.figure(figsize=(8, 5))
sns.histplot(data=df_clean, x='Age', bins=15, kde=True, color='skyblue')
plt.title("Age Distribution of Passengers")
plt.savefig("age_distribution.png")
```
